```python
import bs4 as bs
import urllib.request as url
```

```python
source = url.urlopen('https://www.yelp.com/search?find_desc=&find_loc=San%20Francisco%20Bay%20Area%2C%20CA')
```

```python
page_soup = bs.BeautifulSoup(source, 'html.parser')
```

```python
import csv
csvfile = open('home.csv','w', newline='')
obj = csv.writer(csvfile)
persons=[('Restaurant_name', 'Restaurant_rating', 'Restaurant_noreview', 'Restaurant_address', 'Restaurant_phone', 'Restaurant_review')]
obj.writerows(persons)
major = page_soup.find_all("p", {"class": "lemon--p__373c0__3Qnnj text__373c0__2pB8f text-color--black-regular__373c0__38bRH text-align--left__373c0__2pnx_ text-size--inherit__373c0__2gFQ3"})        
for i in range (len(major)):
    try:
        name = page_soup.find_all("p", {"class": "lemon--p__373c0__3Qnnj text__373c0__2pB8f text-color--black-regular__373c0__38bRH text-align--left__373c0__2pnx_ text-size--inherit__373c0__2gFQ3"})
        name = name[i].text
        print(f"Restaurant_name: {name}")
    except:
        print(None)
    try:
        rating = page_soup.find_all("div", {"class": "lemon--div__373c0__1mboc attribute__373c0__1hPI_ display--inline-block__373c0__2de_K u-space-r1 border-color--default__373c0__2oFDT"})
        rating = rating[1].div.get("aria-label")
        print(f"Restaurant_rating: {rating}")
    except:
        print(None)
    try:
        no_reviews = page_soup.find_all("div", {"class": "lemon--div__373c0__1mboc attribute__373c0__1hPI_ display--inline-block__373c0__2de_K border-color--default__373c0__2oFDT"})            
        no_reviews = no_reviews[i].text
        print(f"Restaurant_noreview: {no_reviews}")
    except:
        print(None)
    try:
        address = page_soup.find_all("address", {"class": "lemon--address__373c0__2sPac"})
        address = address[i].text
        print(f"Restaurant_address: {address}")
    except:
        print(None)
    try:
        phone = page_soup.find_all("div", {"class": "lemon--div__373c0__1mboc container__373c0__19wDx u-padding-l2 border-color--default__373c0__2oFDT text-align--right__373c0__3fmmn"})
        phone = phone[i].text
        phone = ''.join(i for i in phone if i.isdigit())
        print(phone)
    except:
        print(None) 
    try:
        review = page_soup.find_all("p", {"class": "lemon--p__373c0__3Qnnj text__373c0__2pB8f text__373c0__2P1WD alternateStyling__373c0__2ithU text-color--normal__373c0__K_MKN text-align--left__373c0__2pnx_"})
        review = review[i].text
        print(f"Restaurant_review: {review}")
    except:
        print(None) 
    
    persons=[(name, rating, no_reviews, address, phone, review)]
    obj.writerows(persons)
csvfile.close()
```

# Q1 The set of all alphabetic strings

```python
df["Restaurant_review_list"]= ""
```

```python
from nltk import regexp_tokenize
patn = '\w+'
for i in range(len(df)):
    df["Restaurant_review_list"][i] = regexp_tokenize(df["Restaurant_review"][i], patn)
```

# Question 2  The set of all lower case alphabetic strings ending in a b;

```python
df["Restaurant_review_last_b"]= ""
```

```python
for i in range(len(df)):
    list_b = []
    for j in range(len(df['Restaurant_review_list'][i])):
        current_list = df['Restaurant_review_list'][i]
        if not current_list[j][0].isupper():
            if re.search(r"b$", current_list[j]):
                list_b.append(current_list[j])
    df["Restaurant_review_last_b"][i] = list_b
                
            
        
        
        
```
